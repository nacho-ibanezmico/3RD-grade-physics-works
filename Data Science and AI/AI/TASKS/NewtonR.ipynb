```python
import numpy as np
import scipy as sp
import math as mt
from sympy import symbols, sympify, diff


def newton_raphson_sympy(f_str, x0, tolerancia=1e-6, max_iter=100):
    """
    f_str: función como string, ej: "x**2 - 2"
    x0: valor inicial
    """
    x = symbols('x') #symbols = transforms string into variable (symbolic variable)
    f = sympify(f_str) #sympify = tansforms string into function (symbolic expresion)
    f_prime = diff(f, x)  # Calcula la derivada
    x_n = x0

    for i in range(max_iter):
        f_val = float(f.subs(x, x_n)) #Sustituye un símbolo por un valor numérico en una expresión.
        f_prime_val = float(f_prime.subs(x, x_n))

        if abs(f_prime_val) < 1e-10:
            print("Derivada muy pequeña")
            break
        
        x_next = x_n - f_val / f_prime_val
        
        print(f"Aproximación: {x_next}, Iterable: {x_n}, Funcion: {f_val}, DerivFuncion: {f_prime_val}")

        if abs(x_next - x_n) < tolerancia:
            return x_next, i
        
        x_n = x_next

    return x_n, max_iter


raiz, iteraciones = newton_raphson_sympy("x**2 - 2", x0=1)
print(f"Raíz: {raiz}, Iteraciones: {iteraciones}")
```
