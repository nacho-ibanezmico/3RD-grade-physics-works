# Selección hacia delante (Forward Selection)
Forward Selection, es un método de selección de variables para modelos predictivos. Este enfoque comienza con un modelo que no incluye ninguna variable explicativa y luego agrega variables una por una.

En cada paso, se añade la variable que ofrece el mayor incremento en el rendimiento del modelo, hasta que ningún otro incremento significativo es posible o se alcanza el número deseado de variables.

Para este ejemplo, utilizaremos el conjunto de datos de diabetes de Scikit-learn y construiremos un modelo de regresión lineal donde aplicaremos Forward Selection para seleccionar las variables más relevantes.

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import numpy as np
import statsmodels.api as sm
```

```python
# Cargar el conjunto de datos de diabetes
X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python

# Inicialización de variables para Forward Selection
variables = list(range(X_train.shape[1])) # Lista de índices de variables
selected_variables = [] # Variables seleccionadas
performance = [] # Para almacenar el rendimiento del modelo
```

```python
# Forward Selection
for _ in range(len(variables)):
    best_score = np.inf
    best_variable = None
    for variable in variables:
        if variable in selected_variables:
            continue
        # Probar agregar esta variable al modelo
        X_train_selected = sm.add_constant(X_train[:, selected_variables + [variable]])
        model = sm.OLS(y_train, X_train_selected).fit()
        score = model.aic # Usamos el criterio de información de Akaike como métrica de rendimiento

        # Checar si este modelo es el mejor hasta ahora
        if score < best_score:
            best_score = score
            best_variable = variable

    # Si encontramos una nueva mejor variable, la agregamos al modelo
    if best_variable is not None:
        selected_variables.append(best_variable)
        performance.append(best_score)
    else:
        # Si no se encontró una mejora, terminar el loop
        break
```

```python
# Mostrar variables seleccionadas y su rendimiento
selected_variables, performance
```

Este orden indica que la variable 2 fue la primera añadida, ya que su inclusión en el modelo solo proporcionó la mayor mejora en términos de AIC, y así sucesivamente hasta que se agregaron todas las variables. El valor de AIC disminuyó con cada variable agregada, lo que indica una mejora en el ajuste del modelo.

Sin embargo, es importante notar que a medida que agregamos más variables, la mejora en el AIC se vuelve menos significativa, lo que sugiere que algunas de las últimas variables añadidas podrían no contribuir sustancialmente al modelo.

Este proceso de selección nos ayuda a identificar cuáles variables tienen el mayor impacto en la variable objetivo y a construir un modelo que balancee complejidad y rendimiento.

Utilizar Forward Selection es una forma efectiva de mejorar la interpretabilidad del modelo y evitar el sobreajuste, especialmente cuando se trabaja con conjuntos de datos que tienen muchas variables explicativas. ​

### Para entender el código
variables = list(range(X_train.shape[1])): Crea una lista de índices de todas las variables explicativas disponibles.

selected_variables = []: Inicializa una lista vacía para almacenar las variables seleccionadas.

performance = []: Inicializa una lista para almacenar el rendimiento del modelo a medida que se añaden variables.

**Implementación de Forward Selection**:

Inicia un bucle for que potencialmente iterará tantas veces como variables haya.
Dentro del bucle, se inicializa best_score con infinito y best_variable con None. Estas variables almacenarán el mejor puntaje (AIC más bajo) encontrado en esta iteración y la variable que lo proporcionó, respectivamente.

Un segundo bucle for itera sobre todas las variables posibles. Si la variable ya está seleccionada, continúa con la siguiente.

X_train_selected = sm.add_constant(...): Añade una constante (intercepto) al conjunto de características seleccionadas y la nueva variable bajo prueba.

model = sm.OLS(y_train, X_train_selected).fit(): Ajusta un modelo de regresión lineal ordinaria con las variables seleccionadas hasta ahora.

score = model.aic: Obtiene el criterio de información de Akaike del modelo ajustado para evaluar su rendimiento.

Si el score es mejor (más bajo) que best_score, actualiza best_score y best_variable con los valores de esta iteración.

Fuera del segundo bucle, pero aún dentro del primero, si se encontró una best_variable, se añade a selected_variables y su AIC a performance.

Si no, se rompe el bucle principal porque no se encontraron mejoras.

**Salida**:

Finalmente, se imprime la lista de selected_variables y sus respectivos valores de performance (AIC), mostrando el orden en que las variables fueron seleccionadas y cómo afectaron el rendimiento del modelo.
