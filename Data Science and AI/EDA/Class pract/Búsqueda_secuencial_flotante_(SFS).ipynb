# Sequential Floating Searching
Para ilustrar el uso de la Búsqueda Secuencial Flotante (SFS), utilizaremos el conjunto de datos de vino disponible en sklearn.

Este conjunto de datos contiene los resultados de un análisis químico de vinos cultivados en la misma región en Italia pero derivados de tres diferentes tipos de vino (blanco, tinto y rosado).

El análisis determinó las cantidades de 13 componentes encontrados en cada uno de los tres tipos de vinos.

En este ejemplo, nuestro objetivo será seleccionar el subconjunto óptimo de componentes químicos (características) que contribuyen de manera significativa a la clasificación del tipo de vino, utilizando el método de Búsqueda Secuencial Flotante (SFS) implementado en la librería mlxtend.

En este dataset, los componentes del vino son los siguientes:
1. Alcohol: Porcentaje de alcohol en el vino.
2. Ácido málico: Un ácido orgánico encontrado en frutas y vinos.
3. Ceniza: Medida del contenido mineral del vino.
4. Alcalinidad de la ceniza: La alcalinidad de la ceniza, que es la medida de la capacidad de la ceniza para neutralizar el ácido.
5. Magnesio: Nivel de magnesio en el vino, un elemento mineral.
6. Fenoles totales: Cantidad total de compuestos fenólicos en el vino, que son indicativos de la capacidad antioxidante.
7. Flavonoides: Un tipo de fenol, conocidos por sus propiedades antioxidantes y su influencia en el color y sabor del vino.
8. Fenoles no flavonoides: Otro grupo de compuestos fenólicos con efectos antioxidantes.
9. Proantocianidinas: Compuestos polifenólicos que afectan a la astringencia y el color del vino.
10. Intensidad del color: Una medida que indica la profundidad del color del vino.
11. Matiz: Relacionado con el color del vino, pero más específicamente, el matiz o tono.
12. OD280/OD315 de vinos diluidos: Una medida de la absorción de luz que puede estar relacionada con la cantidad de proteínas en el vino.
13. Prolina: Un aminoácido que se encuentra en los vinos, cuya concentración puede variar según el tipo de uva y el lugar de origen.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
# Cargar el conjunto de datos
X, y = load_wine(return_X_y=True)

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

# Normalización de los datos
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(y)
```

```python
# Inicializar el clasificador KNN
knn = KNeighborsClassifier(n_neighbors=3)

# Configurar SFS
sfs = SFS(knn,
          k_features=(1, X.shape[1]),
          forward=True, #Starts with 0 features, and keeps adding 
          floating=True, #Fix up past mistakes
          verbose=2, #will print each update
          scoring='accuracy',
          cv=5) #we reduce overfitting with a k fold splitting the data in 5

# Ejecutar SFS
sfs = sfs.fit(X_train, y_train)
```

```python
# Mostrar las características seleccionadas
selected_features = list(sfs.k_feature_idx_) #INDEX OF THE BEST COLUMNS
print("Características seleccionadas:", selected_features)
```

En este caso, las características que han aparecido en la anterior celda, son:

Característica 0: Alcohol - La cantidad de alcohol en el vino, generalmente en porcentaje.

Característica 1: Ácido málico - Un ácido orgánico presente en los vinos, que influye en el sabor.

Característica 6: Flavonoides - Un tipo de polifenol, sustancias químicas con propiedades antioxidantes, que afectan al color y sabor del vino.

Característica 7: Fenoles no flavonoides - Otro tipo de polifenol, que también contribuyen a las propiedades antioxidantes del vino.

Característica 9: Color Intensidad - La intensidad del color del vino, que puede dar pistas sobre la edad y la concentración de ciertos compuestos.

Característica 12: Prolina - Un aminoácido que se encuentra en los vinos, su concentración puede estar relacionada con el lugar de origen y el tipo de uva.

Estos componentes químicos son fundamentales para determinar las propiedades organolépticas (sabor, aroma, color) del vino, así como su calidad y potencial de envejecimiento. La selección de estas características sugiere que son factores importantes para distinguir entre los diferentes tipos de vino en el conjunto de datos, lo que puede reflejar diferencias en el proceso de vinificación o en las variedades de uva utilizadas.
