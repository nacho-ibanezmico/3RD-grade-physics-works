#Recursive Feature Elimination (RFE) con SVM
Supongamos que queremos seleccionar las características más relevantes de un conjunto de datos para la predicción de cáncer usando un modelo SVM (Support Vector Machine). Utilizaremos el conjunto de datos de cáncer de mama de Wisconsin (disponible en sklearn.datasets) para demostrar cómo implementar RFE con SVM.

Tranquilos/as veremos el funcionamiento del SVM en la asignatura de Aprendizaje Automático.  

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Cargar el conjunto de datos
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
# Crear el modelo SVM
svc = SVC(kernel="linear")
```

```python
# Crear el selector RFE utilizando el modelo SVM como estimador
rfe = RFE(estimator=svc, n_features_to_select=10, step=1)
rfe = rfe.fit(X_train, y_train)
```

```python
# Predecir utilizando el modelo entrenado y el conjunto de prueba
y_pred = rfe.predict(X_test)
```

```python
# Calcular y mostrar la precisión del modelo
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

```python
# Mostrar las características seleccionadas
print(f"Características seleccionadas: {data.feature_names[rfe.support_]}")
```

Este código carga el conjunto de datos de cáncer de mama, divide los datos en conjuntos de entrenamiento y prueba, utiliza RFE con un modelo SVM para seleccionar las 10 características más relevantes y, finalmente, evalúa el modelo en el conjunto de prueba.

El resultado incluye la precisión del modelo junto con las características seleccionadas por RFE, proporcionando así una visión sobre qué características son más importantes para la predicción de cáncer de mama en este contexto.
