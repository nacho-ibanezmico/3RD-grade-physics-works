```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
# DEFINIMOS LAS CONSTANTES DEL PROBLEMA:
h = 1
m = 1
L = 1
N = 100 #Puntos que quiero que tenga el sistema entre 0 y L

x = np.linspace(0, L, N+1) #DISCRETIZO
n_leves = np.arange(1, 5)

# DEFINIMOS LA FUNCIÓN DE ONDA:
def psi(n, x):
    return np.sqrt(2/L) * np.sin(((n * np.pi * x) / L))

def ener(n,h,m,L):
    return (np.pi**2 * h**2 * n**2)/(2*m*L**2)

for i in n_leves:
    func = [psi(i, j) for j in x]
    En = ener(i, h, m, L)  
    print(f"n={i}, E={En:.6f}")
    plt.scatter(x, func, s=8, label=f"n={i}")
plt.xlabel("x")
plt.ylabel(r"$\psi_n(x)$")
plt.legend()
plt.show()
```
